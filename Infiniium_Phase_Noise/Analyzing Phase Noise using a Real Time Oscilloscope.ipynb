Overview:
A Real-Time sampling oscilloscope is designed to capture time-domain phenomena but the combination of an extremely accurate timebase and Fast Fourier Transforms can make them very useful at measuring frequency domain phenomena also.
Phase Noise is traditionally measured using a frequency domain instrument such as a Spectrum Analyzer or specialised Signal Source Analyzer. While these frequency domain instruments typically have high dynamic range there are limitations to their measurement capability in terms of the offset frequency they are capable of measuring to or the fact they are unable to measure the phase noise of a non-clock signal.

This note will explain how to measure the phase noise of a clock or data signal using an Agilent Infiniium real-time sampling oscilloscope using a tool called Infiniium Phase Noise.

```python
f = 1e9 # The sinusoid base frequency 1GHz
sa_rate = 80e9
mem_depth = 2**21
t = linspace(0,mem_depth/sa_rate,mem_depth)
dv = 0 # Amplitude Modulation, AM
dt = 0 # Phase Modulation, PM
v = (1+dv)*sin(2*pi*f*t + dt)
v_spec = fft.fft(v)
freqs = fft.fftfreq(mem_depth, 1/sa_rate)
figure(figsize=(15,4))
subplot(1,2,1)
plot(t[:2000],v[:2000])
subplot(1,2,2)
semilogy(freqs[:len(v_spec)/32],abs(v_spec[:len(v_spec)/32])/len(v_spec)**2)
```

```python
# Add some sinusoidal AM @ 100MHz
dv = 0.1*(sin(2*pi*0.1*f*t))
dt = 0
v = (1+dv)*sin(2*pi*f*t + dt)
v_spec = fft.fft(v)
freqs = fft.fftfreq(mem_depth, 1/sa_rate)
figure(figsize=(15,4))
subplot(1,2,1)
plot(t[:2000],v[:2000])
subplot(1,2,2)
semilogy(freqs[:len(v_spec)/32],abs(v_spec[:len(v_spec)/32])/len(v_spec)**2)
```

```python
# Add some sinusoidal PM @ 100MHz
dv = 0
dt = 0.1*sin(2*pi*0.1*f*t)
v = (1+dv)*sin(2*pi*f*t + dt)
v_spec = fft.fft(v)
freqs = fft.fftfreq(mem_depth, 1/sa_rate)
figure(figsize=(15,4))
subplot(1,2,1)
plot(t[:4000],v[:4000], t[:4000],dt[:4000])
subplot(1,2,2)
semilogy(freqs[:len(v_spec)/32],abs(v_spec[:len(v_spec)/32])/len(v_spec)**2)
```
